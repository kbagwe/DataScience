```python
#4.[5pts] Create a new dataframe, df2, that contains only the following fields 'Vehicle_Anti_Theft_Device', 'Driver_Minimum_Age', 'Vehicle_Age_In_Years',  'Annual_Premium', 'Loss_Amount'. Note that 'PolicyNo' is the index of this dataframe as well. To answer the following questions use df2.
dataset= 'example_portfolio.csv'
df2 = pd.read_csv(dataset,index_col=0,usecols=['PolicyNo','Vehicle_Anti_Theft_Device','Driver_Minimum_Age','Vehicle_Age_In_Years','Annual_Premium','Loss_Amount'])
#usecolslist-like or callable, default None
#Return a subset of the columns. If list-like, all elements must either be positional (i.e. integer indices into the document columns) or strings that correspond to column names provided either by the user in names or inferred from the document header row(s). For example, a valid list-like usecols parameter would be [0, 1, 2] or ['foo', 'bar', 'baz'].
#index_colint, str, sequence of int / str, or False, default None
#Column(s) to use as the row labels of the DataFrame, either given as string name or column index. If a sequence of int / str is given, a MultiIndex is used.
```

```python
#5.[5pts] Show the mean, std, min, and max for the numerical fields.
for col in df2.select_dtypes(include=['int64','float64']):
    print('Name of column : ' + col)
    print('mean :%f ' % df2[col].mean())
    print('std :%f ' % df2[col].std())
    print('min :%f ' % df2[col].min())
    print('max :%f ' % df2[col].max())
#DataFrame.select_dtypes(include=None, exclude=None)
#Return a subset of the DataFrameâ€™s columns based on the column dtypes.
```

```python
#6.[5pts] Show the unique values of 'Vehicle_Anti_Theft_Device'.
print("The Unique values of Vehicle_Anti_Theft_Device :")
for col1 in df2['Vehicle_Anti_Theft_Device'].unique():
    print(col1)
#.unique() provides unique values inside a dataframe columns
```

```python
# (7) [5pts] Calculate and print out the Portfolio Premium, Portfolio Loss, and Loss Ratio. Portfolio Premium = sum of all the Annual Premiums of all the policies. Portfolio Loss = sum of all the Loss Amounts of all the policies. Loss Ratio = Portfolio Loss / Portfolio Premium.
#Portfolio Premium
Portfolio_Premium = df2['Annual_Premium'].sum()
print('Portfolio Premium:%f' % Portfolio_Premium)
#.sum() provides sum of that columns
#Portfolio Loss 
Portfolio_Loss = df2['Loss_Amount'].sum()
print('Portfolio Loss:%f' % Portfolio_Loss)
##.sum() provides sum of that columns
#Loss Ratio 
Loss_Ratio = Portfolio_Loss / Portfolio_Premium
print('Loss Ratio:%f' % Loss_Ratio )
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
%matplotlib inline

#9.[5pts] Create a new data frame, df3, that groups the rows by Vehicle_Anti_Theft_Device and aggregates them by taking the mean for all the fields. Note that the index of the new data frame should be Vehicle_Anti_Theft_Device. To answer the following questions use df3.
dataset= 'example_portfolio.csv'
df = pd.read_csv(dataset)
df3 = df.groupby('Vehicle_Anti_Theft_Device').mean()
# groupby() Group DataFrame using a mapper or by a Series of columns.
#A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.
#mean() function can be used to calculate mean/average of a given list of numbers. It returns mean of the data set passed as parameters.
```

```python
#10. [5pts] Display the index of the new data frame.
print(df3.index)
#The index() method returns the position at the first occurrence of the specified value.
```

```python
# 11.[5pts] Create a new field 'Loss_Ratio' that gives the loss ratio per type of vehicle anti theft device. Loss Ratio = Loss_Amount / Annual_Premium.
df3['Loss_Ratio'] = df3['Loss_Amount'] / df3['Annual_Premium']
```

```python
#12.[5pts] Sort the data frame by Loss Ratio in ascending order.
df3.sort_values(by = 'Loss_Ratio')
#sort_values() - Sort by the values along either axis
```

```python
#13.[5pts] Create a bar plot for 'Vehicle_Anti_Theft_Device' vs 'Loss_Ratio'.
df3['Loss_Ratio'].plot.bar()
#A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent. A bar plot shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.
```
